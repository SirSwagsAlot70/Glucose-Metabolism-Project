```python
# Imports
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
def Glucose_Met(t,y,K_cat_f,K_cat_r,K_m_f,K_m_r,C_G,K_G,C_I,K_I,A_G,A_I):    
    #y[0] = S
    #y[1] = P
    #y[2] = G
    #y[3] = I
    
    dydt = np.zeros(4)

    V = ((K_cat_f*y[2]*y[0]/K_m_f)-(K_cat_r*y[3]*y[1])/K_m_r)/(1+(y[0]/K_m_f)+(y[1]/K_m_r))
    F_G = (C_G/(1+(y[1]/K_G)))
    F_I = ((C_I*(y[1]/K_I))/(1+(y[1]/K_I)))
    
    dydt[0] = -1*V
    dydt[1] = V
    dydt[2] = F_G - A_G*y[2]
    dydt[3] = F_I - A_I*y[3]
    
    return dydt
```

```python
# Initial Conditions
y0 = [5,0,0,0]
# time span
t = np.linspace(0,100,1000)
tspan = [t[0],t[-1]]
K_cat_f = 1
K_cat_r = 1
K_m_f = 1
K_m_r = 1
C_G = 1
K_G = 1
C_I = 1
K_I = 1
A_G = 1
A_I = 1

ode_sol = solve_ivp(lambda t,y:Glucose_Met(t,y,K_cat_f,K_cat_r,K_m_f,K_m_r,C_G,K_G,C_I,K_I,A_G,A_I),tspan,y0,t_eval=t)

plt.plot(t,ode_sol.y[0],'r')
plt.plot(t,ode_sol.y[1],'orange')
plt.plot(t,ode_sol.y[2],'y')
plt.plot(t,ode_sol.y[3],'g')

plt.xlabel('Healthy Model, time [s]')
plt.ylabel('concentration [mM]')
plt.legend(['Glycogen','Glucose','GLucagon','Insulin']);
```
